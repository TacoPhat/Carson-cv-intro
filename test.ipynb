```python
img = cv2.imread('pool_lane.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to grayscale
#plt.imshow(gray)
num = 27
edges = cv2.Canny(gray, num, num + 5, apertureSize=3) # detect edges
#plt.imshow(edges)
lines = cv2.HoughLinesP(
                edges,
                1,
                np.pi/180,
                10,
                minLineLength=25,
                maxLineGap=20,
) # detect lines
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

#print(lines)


plt.imshow(img)

```
